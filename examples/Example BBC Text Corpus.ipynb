# Example BBC Clustering
This is an example for the clustering using preset parameters on a large text corpus. The goal is to cluster the dataset and extract persitent topics. 
## Import Section

```python
import pandas as pd 
from clustering_pipeline import ClusteringPipeline
import matplotlib.pyplot as plt
from importlib import resources
```

## Load Dataframe

```python

PACKAGE_NAME = 'tactic.example_data'
RESOURCE_NAME = 'bbc-text.csv'


with resources.files(PACKAGE_NAME).joinpath(RESOURCE_NAME).open('r') as f:
    df = pd.read_csv(f, , encoding="latin1")

print(f"Loaded {len(df)} records")
print(f"Columns: {df.columns.tolist()}")
```

## Run Clustering 
Clustering is ran with fixed, pretrained parameters. 


```python
pipeline = ClusteringPipeline(df, text_column='text')

results = pipeline.full_pipeline(
        tune_hyperparameters=True,   # enables tuning instead of fixed params
        visualize=True,              # produce t-SNE visualization
        extract_keywords=True,       # extract keywords after clustering
        analyze_topics=True,         # run LDA topic modeling
        min_clusters=10,              # minimum cluster count requirement
        time_limit=1000,              # tuning time in seconds
        standardize=True,            # scale data before clustering
        tune_tsne=False,             # skip t-SNE tuning
        random_state=42,             # ensure reproducibility
        num_topics=10,               # number of LDA topics
        passes=15,                   # number of LDA passes
        tf_top_n=5,                  # number of TF keywords per cluster
        yake_top_n=10,               # YAKE initial keywords
        yake_final_n=5,              # YAKE final filtered keywords
    )

# Access results
print("\n=== PIPELINE RESULTS ===")
print("-" * 60)

print("\nCluster Summary:")
print(results['dataframe']['Clusters'].value_counts())

if results.get('keywords') is not None:
    print("\nTop Keywords per Cluster:")
    print(results['keywords'].head(10))

if results.get('topics') and 'cluster_topics' in results['topics']:
    print("\nTopics by Cluster:")
    for cluster_id, topic_list in results['topics']['cluster_topics'].items():
        print(f"\nCluster {cluster_id}:")
        for topic_words, weight in topic_list[:3]:
            print(f"  - {topic_words} (weight: {weight:.3f})")

print("\n=== Pipeline complete ===")
```

### Keyword Extraction Results

```python
import pandas as pd

if results.get('topics') and 'cluster_topics' in results['topics']:
    print("\nTopics by Cluster with Keywords:")

    cluster_topics = results['topics']['cluster_topics']
    keywords_df = results.get('keywords')

    for cluster_id, topic_ids in cluster_topics.items():
        print(f"\nCluster {cluster_id}:")
        print(f"  Assigned topics: {topic_ids}")

        if isinstance(keywords_df, pd.DataFrame):
            # Look up the corresponding keywords for this cluster
            row = keywords_df[keywords_df['cluster'] == cluster_id]

            if not row.empty:
                yake_short = row['Yake Short'].values[0]
                yake_long = row['Yake Long'].values[0]
                tfidf = row['TFIDF'].values[0]
                tf = row['TF'].values[0]

                print("  Top Keywords:")
                print(f"    - YAKE (Short): {yake_short}")
                print(f"    - YAKE (Long):  {yake_long}")
                print(f"    - TF:            {tf}")
                print(f"    - TFIDF:         {tfidf}")
            else:
                print("  - No keywords available for this cluster.")
        else:
            print("  - No keyword data available.")
```
